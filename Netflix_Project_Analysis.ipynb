```python
#Task 5: In which year were the highest number of TV shows and movies released? Display with Bar chart.
olliedata['Date_Convert'] = pd.to_datetime(olliedata['Release_Date'])           #Need to create new date column to compare
olliedata.head()
```

```python
counts_year = olliedata['Date_Convert'].dt.year.value_counts()            #Counts the occurrence of each individuals in Date column
counts_year.index = counts_year.index.astype(int)                         #Change the year display to eliminate decimal (like 2019.0 ; 2020.0) -----
counts_year
```

```python
#Graph a Bar chart for visualization
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))  # Set the size of the figure (width, height)
bar_chart = counts_year.plot(kind = 'bar',color='skyblue', edgecolor='black', linewidth=1.5)


# Customizations
plt.xlabel('Year', fontsize=14)                                                         # Add label to the x-axis
plt.ylabel('Number of TV Shows & Movies', fontsize=14)                                  # Add label to the y-axis
plt.title('Highest TV Shows & Movies Released Each Year', color= 'red', fontsize=16)    # Add a title to the chart

# Add data labels on top of each bar
for bar in bar_chart.patches:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(height),
             ha='center', va='bottom', fontsize=12)


# Customize the appearance
plt.xticks(rotation=45)                        # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.5)  # Add grid lines for the y-axis
plt.tight_layout()                             # Adjust the layout to prevent labels from being cut off

# Show the plot
plt.show()
```

```python
#Task 9: Show total number of TV Show over the years via graphs. Show all the TV Show which has equal or greater than 5 Seasons.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(11, 7))

# Filter the DataFrame to get only 'TV Show' categories
cate_tvshow = olliedata[olliedata['Category'] == 'TV Show']           #List of only 'TV show'

# Calculate the count of 'TV Show' occurrences over the years
category_s1 = cate_tvshow['CYear'].value_counts().sort_index()
category_s1.index = category_s1.index.astype(int)  

# Create a line plot
sns.lineplot(x= category_s1.index, y=category_s1, color = 'black')


# Visuals enhancement
for x, y in zip(category_s1.index, category_s1):
    # Add a bounding box around the data label
    bbox_props = dict(boxstyle='round,pad=0.3', fc='lightgray', alpha=0.7)

    # Annotate the data point with the data label
    plt.annotate(f'{y}', xy=(x, y), xytext=(0,7), textcoords='offset points',
                 bbox=bbox_props, ha='center', fontsize=10, va='bottom')



plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.title("Total 'TV Show' Over the Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()

```

```python
#Task 13: Show the top 5 countries with Highest number of TV Shows By Graph + Highlight the highest number
new_data.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#Count number of TV shows for each country
tv_show_counts = new_data[new_data['Category'] == 'Movie']['Country'].value_counts().nlargest(5)
tv_show_counts
```

```python
# Plot the horizontal bar graph
plt.figure(figsize=(10, 7))
bars = plt.barh(tv_show_counts.index, tv_show_counts.values, color='midnightblue')


# Find the index of the highest country and set a distinct color for it
highest_country = tv_show_counts.idxmax()
highest_country_index = tv_show_counts.index.get_loc(highest_country)
bars[highest_country_index].set_color('darkorange')

plt.xlabel('Number of TV Shows')
plt.ylabel('Country')
plt.title('Top 5 Countries with the Highest Number of TV Shows')

# Add data labels to the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2, f'{int(width)}', ha='left', va='center', fontsize=12, color='black')

plt.gca().invert_yaxis()  # Invert the y-axis to display countries with the highest count on top
plt.tight_layout()
plt.show()
```
